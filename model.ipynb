```python

# Function to load and preprocess the dataset
def load_dataset(dataset_path, image_width, image_height):
    images = []  # Used to store image data
    labels = []  # Used to store class labels
    label_encoder = LabelEncoder()

    for class_name in os.listdir(dataset_path):
        class_folder = os.path.join(dataset_path, class_name)
        for image_file in os.listdir(class_folder):
            image_path = os.path.join(class_folder, image_file)
           
            # Load and preprocess the image
            image = cv2.imread(image_path)
            image = cv2.resize(image, (image_width, image_height))
           
            # Append image and label
            images.append(image)
            labels.append(class_name)

    # Encode class names to numerical labels
    labels = label_encoder.fit_transform(labels)

    return np.array(images), to_categorical(labels, num_classes=num_classes)

# Load and preprocess your dataset
images, labels = load_dataset(dataset_path, image_width, image_height)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(images, labels, test_size=0.2, random_state=42)

# Define your model
model = Sequential()

# Add Conv2D and MaxPooling2D layers for feature extraction from images
model.add(Conv2D(32, kernel_size=(3, 3), input_shape=(image_width, image_height, num_channels), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(64, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

# Flatten the output from convolutional layers
model.add(Flatten())

# Add LSTM layers
model.add(LSTM(128, return_sequences=True))
model.add(LSTM(64, return_sequences=False))

# Add fully connected layers
model.add(Dense(256, activation='sigmoid'))
model.add(Dense(num_classes, activation='softmax'))

# Compile your model with appropriate loss, optimizer, and metrics
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Print the model summary
model.summary()

# Train your model
history = model.fit(X_train, y_train, batch_size=32, epochs=10, validation_data=(X_test, y_test))

# Evaluate your model
test_loss, test_acc = model.evaluate(X_test, y_test)
print(f"Test accuracy: {test_acc * 100:.2f}%")
```

```python
import cv2
import os

# Define the input directory where your videos are located
input_directory = r'D:\DATA'  # Replace with your actual path

# Define the output directory for saving extracted frames
output_directory = r'C:\Users\soham\python prog\DATA2t'  # Replace with your desired output path

# Set the frame rate (1 fps)
fps = 1

# Loop through the subfolders in the input directory
for class_folder in os.listdir(input_directory):
    class_path = os.path.join(input_directory, class_folder)
    
    # Check if the item in the folder is a subfolder (class folder)
    if os.path.isdir(class_path):
        # Create an output directory for this class within the specified output directory
        class_output_dir = os.path.join(output_directory, class_folder)
        os.makedirs(class_output_dir, exist_ok=True)
        
        # Loop through video files in the class folder
        for video_filename in os.listdir(class_path):
            if video_filename.endswith('.mp4'):
                video_path = os.path.join(class_path, video_filename)
                cap = cv2.VideoCapture(video_path)
                
                frame_count = 0
                time_between_frames = int(cap.get(cv2.CAP_PROP_FPS) / fps)
                
                while True:
                    ret, frame = cap.read()
                    
                    if not ret:
                        break
                    
                    if frame_count % time_between_frames == 0:
                        # Save the frame as an image in the class's output directory
                        image_filename = os.path.join(class_output_dir, f"{video_filename}_{frame_count}.jpg")
                        cv2.imwrite(image_filename, frame)
                    
                    frame_count += 1
                
                # Release the video capture object
                cap.release()

# All frames from videos in different classes should now be organized in respective class folders within the 'C:\Users\asims\Desktop\output' directory, with one frame per second.
```
